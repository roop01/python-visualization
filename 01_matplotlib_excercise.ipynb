# Matplotlib Excercise

**Problem 1.1**

Draw with marker and annotate the plot

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 5.0, 0.1)
y = np.cos(2 * np.pi * x) * np.exp(-x)

plt.subplot()
ax = plt.gca()
ax.set_facecolor("pink")
plt.plot(x, y, marker='o', mfc = 'c',  mec = 'c', c='k')
plt.title('Plot 1')
plt.xlabel('Abscissa', color='green')
plt.ylabel('Ordinate', color='green')
plt.annotate('2nd crest', xy = (2, .17), xytext = (2, .4), arrowprops = dict(facecolor ='green',
                                  shrink = 1.5))
plt.annotate('θ = 60°', xy = (1, .4), xytext = (.9, .43) )
plt.grid(True)
plt.show()
```

**Problem 1.2**

Multiple subplots

```python
"""
Simple demo with multiple subplots.

"""
import numpy as np
import matplotlib.pyplot as plt


x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, marker='o', mfc = 'r',  mec = 'c')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, '.-' , mfc = 'r',  mec = 'c')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```
